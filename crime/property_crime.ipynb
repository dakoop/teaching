```python
import pandas as pd
df = pd.read_csv("crime.csv", header=None)
```

```python
df.columns = ["year", "crime_rate"]
```

```python
df.head(10)
```

```python
df.dropna(axis=0, how='all', inplace=True)
```

```python
df.head(10)
```

```python
index = df.year.str.contains("Reported")
```

```python
df.year = df.year.str.replace("Reported crime in ", "")
```

```python
df.head(10)
```

```python
index = ~df.year.str.contains("\d+")
```

```python
df["state"] = df[index].year
```

```python
df.head(10)
```

```python
df.state.fillna(method="pad",axis=0,inplace=True)
df.head(10)
```

```python
df.dropna(axis=0,inplace=True)
```
